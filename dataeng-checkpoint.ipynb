```python
def safe_get(d, path, default=None):
    """Acceso seguro a diccionarios anidados."""
    try:
        for key in path:
            d = d[key]
        return d
    except:
        return default


class Match:
    def __init__(self, gameId, homeTeamId, awayTeamId, homeTeamName, awayTeamName,
        homeTeamAbbr, awayTeamAbbr, leagueName, leagueAbbr, homeTeamScore, awayTeamScore,
        attendance, venueId, fullName, city, country,fecha, referee,stage):

        self.gameId = gameId
        self.homeTeamId = homeTeamId
        self.awayTeamId = awayTeamId
        self.homeTeamName = homeTeamName
        self.awayTeamName = awayTeamName
        self.homeTeamAbbr = homeTeamAbbr
        self.awayTeamAbbr = awayTeamAbbr
        self.leagueName = leagueName
        self.leagueAbbr = leagueAbbr
        self.homeTeamScore = homeTeamScore
        self.awayTeamScore = awayTeamScore
        self.attendance = attendance
        self.venueId = venueId
        self.fullName = fullName
        self.city = city
        self.country = country
        self.fecha = fecha
        self.referee = referee
        self.stage= stage

    def to_dict(self):
        return self.__dict__

def empty_match(gameId):
    return {
        "gameId": gameId,
        "homeTeamId": None,
        "awayTeamId": None,
        "homeTeamName": None,
        "awayTeamName": None,
        "homeTeamAbbr": None,
        "awayTeamAbbr": None,
        "leagueName": None,
        "leagueAbbr": None,
        "homeTeamScore": None,
        "awayTeamScore": None,
        "attendance": None,
        "venueId": None,
        "fullName": None,
        "city": None,
        "country": None,
        "fecha": None,
        "referee": None,
        "stage": None
    }


def extraer(id,max_retries=50):
    url = (
            "https://site.web.api.espn.com/apis/site/v2/sports/soccer/mex.1/summary"
            "?region=us&lang=en&contentorigin=espn&event=" + str(id)
        )

    response = requests.get(url, timeout=5)

    if response.status_code == 403:
        raise RuntimeError("RATE_LIMIT")

    if response.status_code == 404:
        return empty_match(id)

    if response.status_code != 200:
        return None

    data = response.json()

    home_id = int(safe_get(data, ["header","competitions",0,"competitors",0,"team","id"], 0))
    away_id = int(safe_get(data, ["header","competitions",0,"competitors",1,"team","id"], 0))

    home_name = safe_get(data, ["header","competitions",0,"competitors",0,"team","displayName"], None)
    away_name = safe_get(data, ["header","competitions",0,"competitors",1,"team","displayName"], None)

        # Validación mínima
    if home_id == 0 or away_id == 0 or home_name is None or away_name is None:
        return empty_match(id)

    partido = Match(
            gameId=id,
            homeTeamId=home_id,
            awayTeamId=away_id,
            homeTeamName=home_name,
            awayTeamName=away_name,
            homeTeamAbbr=safe_get(data, ["header","competitions",0,"competitors",0,"team","abbreviation"], None),
            awayTeamAbbr=safe_get(data, ["header","competitions",0,"competitors",1,"team","abbreviation"], None),
            leagueName=safe_get(data, ["header","league","name"], None),
            leagueAbbr=safe_get(data, ["header","league","abbreviation"], None),
            homeTeamScore=int(safe_get(data, ["header","competitions",0,"competitors",0,"score"], 0)),
            awayTeamScore=int(safe_get(data, ["header","competitions",0,"competitors",1,"score"], 0)),
            attendance=safe_get(data, ["gameInfo", "attendance"], None),
            venueId=int(safe_get(data, ["gameInfo", "venue", "id"], 0)),
            fullName=safe_get(data, ["gameInfo", "venue", "fullName"], None),
            city=safe_get(data, ["gameInfo", "venue", "address", "city"], None),
            country=safe_get(data, ["gameInfo", "venue", "address", "country"], None),
            fecha=safe_get(data, ["header","competitions",0,"date"], None),
            referee=safe_get(data, ["gameInfo", "officials", 0, "fullName"], None),
            stage=safe_get(data,["header","season","name"],None)
        )

    return partido.to_dict()


def descargar_masivo(desde, hasta, xxx, workers=200):
    ids = [i for i in range(desde, hasta + 1) if i in xxx] # agregar el not
    resultados = []
    i = 0
    print(f"Descargando {len(ids)} partidos con {workers} threads...")
    while i < len(ids):

        with ThreadPoolExecutor(max_workers=workers) as executor:
            #futures = {executor.submit(extraer, i): i for i in ids}
            futures = {
                executor.submit(extraer, ids[j]): j
                for j in range(i, min(i + workers, len(ids)))
            }

            for future in as_completed(futures):
                
                idx = futures[future]
                current_id = ids[idx]

                try:
                    r = future.result()
                    if r is not None:
                        resultados.append(r)

                    i += 1

                except RuntimeError as e:

                    if str(e) == "RATE_LIMIT":
                        print("RATE LIMIT DETECTADO — esperando 60 segundos...\n")
                        time.sleep(120)
                        break
                
    return resultados


def juntar(lista):
    merged = defaultdict(list)
    for d in lista:
        for k, v in d.items():
            merged[k].append(v)
    return merged
```
